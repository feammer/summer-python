# tkinter 界面

```python
from tkinter import *
from PIL import Image

def gui_label(root_ui):
    customAttribute = {'anchor':'w','font':'Arial,28'}
    user_label = Label(master=root_ui,text='账号',cnf=customAttribute)
    user_label.place(x=20,y=40,width = 100, height = 20)
    password_label = Label(master=root_ui,text='密码',cnf=customAttribute)
    password_label.place(x=20,y=80,width = 100, height = 20)

def gui_textbox(root_ui):
    user_textbox = Entry(master=root_ui)
    user_textbox.place(x=120,y=40,width = 160, height = 20)
    password_textbox = Entry(master=root_ui)
    password_textbox.place(x=120, y=80, width = 160, height = 20)

def gui_button(root_ui):
    def login_action():
        img = Image.open('test.jpg')
        img.show()
    def exit_action():
        root_ui.destroy()
    login_button = Button(master=root_ui,text='登录',command=login_action)
    login_button.place(x=80, y=160, width = 60, height = 20)
    exit_button = Button(master=root_ui,text='取消',command=exit_action)
    exit_button.place(x=180, y=160, width = 60, height = 20)

def gui_start():
    login = Tk()
    login.title('登录')
    login.geometry('320x240')
    gui_label(login)
    gui_textbox(login)
    gui_button(login)
    login.mainloop()

gui_start()
```

# matplotlib 绘制心形线

```python
from numpy import linspace, sin, cos, power, pi
import matplotlib.pyplot as plt

t = linspace(0, pi, 1000)
x = sin(t)
y = cos(t) + power(x, 2.0 / 3)
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.plot(x, y, color='red')
plt.plot(-x, y, color='red')
plt.savefig('心形线.jpg')
```

# Women 点线图

## 使用 pandas

```python
import pandas as pd

women = pd.read_csv('women.csv')
women.plot(x='height', y='weight')
plt.savefig('women点线图-pandas.jpg')
```

## 使用 matplotlib

```python
plt.plot(women['height'], women['weight'])
plt.xlabel('height')
plt.ylabel('weight')
plt.savefig('women点线图-matplotlib.jpg')
```

## 使用 seaborn

```python
from seaborn import lineplot

lineplot(x='height', y='weight', data=women)
plt.savefig('women点线图-seaborn.jpg')
```
